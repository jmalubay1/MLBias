```python
import os
import numpy as np
import pandas as pd  # For reading CSV files
import time
import matplotlib.pyplot as plt

# Path to all CSVs
DATA_PATH = 'compas-scores-two-years.csv'
```

```python
data = pd.read_csv(DATA_PATH)
```

```python
def pie_chart(column):
    labels = data[column].unique()
    dist = data[column].value_counts()

    # plot it
    fig1, ax1 = plt.subplots()
    ax1.pie(dist, labels=labels, autopct='%1.1f%%', startangle=90)
    ax1.axis('equal')
    plt.show()

def bar_graph(column):
    labels = data[column].unique()
    dist = data[column].value_counts()

    
    # plot it
    fig2, ax2 = plt.subplots()
    y_pos = np.arange(len(labels))
    ax2.bar(y_pos, dist, align='center', alpha=0.5)
    plt.xticks(y_pos, labels)
    plt.show()
```

```python
pie_chart('race')
```

```python
pie_chart('sex')
bar_graph('sex')
```

```python

bar_graph('c_charge_degree')
bar_graph('decile_score')
```
